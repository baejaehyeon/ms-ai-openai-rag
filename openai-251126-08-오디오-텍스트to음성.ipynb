```python
from openai import AzureOpenAI
from dotenv import load_dotenv
import os

# .env 환경변수 로드
load_dotenv()
AZURE_CHRIS_KEY = os.getenv("AZURE_CHRIS_KEY")

client = AzureOpenAI(
    azure_endpoint='https://8a000-openai.openai.azure.com/',
    api_key=AZURE_CHRIS_KEY,
    api_version="2025-03-01-preview",
)

prompt = "안녕하세요. 시각 장애인을 위한, The Mind's Eyes 입니다. 생성을 원하시는 이미지에 대해 설명해주세요."

response = client.audio.speech.create(
    input = prompt,
    model = 'gpt-4o-mini-tts',
    voice= 'shimmer'
)
 
response.write_to_file('files/guide.mp3')
```
