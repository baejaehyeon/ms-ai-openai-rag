```python
import os
from openai import AzureOpenAI

# 마이크 입력
import sounddevice as sd
from scipy.io.wavfile import write

from dotenv import load_dotenv

# .env 환경변수 로드
load_dotenv()
AZURE_CHRIS_KEY = os.getenv("AZURE_CHRIS_KEY")
    
client = AzureOpenAI(
    api_key=AZURE_CHRIS_KEY,
    api_version="2024-06-01",
    azure_endpoint = 'https://8a000-openai.openai.azure.com/'
)

second = 5 # 녹음 시간
audio_file_path = 'files/output.wav'

print("녹음 시작...")
audio = sd.rec(int(5 * 16000), samplerate=16000, channels=1, dtype='int16')
sd.wait()  # 녹음 끝날 때까지 대기
write(audio_file_path, 16000, audio)
print(f"녹음 완료 → '{audio_file_path}' 저장됨")

with open(audio_file_path, 'rb') as audio_file:
    transcription = client.audio.transcriptions.create(
        model = 'whisper',
        language="ko",
        file = audio_file
    )
 
transcription
```
