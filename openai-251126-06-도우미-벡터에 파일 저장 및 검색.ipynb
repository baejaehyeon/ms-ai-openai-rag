#### 벡터에 저장된 파일 분석

```python
import os
import json
import requests
import time
from openai import AzureOpenAI

client = AzureOpenAI(
  azure_endpoint = os.getenv("AZURE_OAI_ENDPOINT"),
  api_key= os.getenv("AZURE_OAI_KEY"),
  api_version="2024-05-01-preview"
)

# 도우미 객체 생성(Azure OpenAI 벡터 저장소 ID : vs_LLnqkbYWTf1rCbvdsM789HzP)
assistant = client.beta.assistants.create(
  model="gpt-4o-mini", # replace with model deployment name.
  instructions="",
  tools=[{"type":"file_search"}],
  tool_resources={"file_search":{"vector_store_ids":["vs_LLnqkbYWTf1rCbvdsM789HzP"]}},
  temperature=1,
  top_p=1
)

# Create a thread
thread = client.beta.threads.create()

# Add a user question to the thread
message = client.beta.threads.messages.create(
  thread_id=thread.id,
  role="user",
  content="백터 저장소에 저장된 파일 분석해줘" # Replace this with your prompt
)



# Run the thread
run = client.beta.threads.runs.create(
  thread_id=thread.id,
  assistant_id=assistant.id
)

# Looping until the run completes or fails
while run.status in ['queued', 'in_progress', 'cancelling']:
  time.sleep(1)
  run = client.beta.threads.runs.retrieve(
    thread_id=thread.id,
    run_id=run.id
  )

if run.status == 'completed':
  messages = client.beta.threads.messages.list(
    thread_id=thread.id
  )
  #print(messages.json())
  print(messages.data[0].content[0].text.value)
elif run.status == 'requires_action':
  # the assistant requires calling some functions
  # and submit the tool outputs back to the run
  pass
else:
  print(run.status)



```
