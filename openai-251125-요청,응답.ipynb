```python
# 요청에 필요한 모듈
import requests
from dotenv import load_dotenv
import os

# .env 환경변수 로드
load_dotenv()
AZURE_OAI_KEY = os.getenv("AZURE_OAI_KEY")

endpoint = "https://8ai022-openai.openai.azure.com/openai/deployments/gpt-4o-mini/chat/completions?api-version=2025-01-01-preview"
headers = {
    'api-key' : AZURE_OAI_KEY,
    'Content-Type' : 'application/json'
}

def request_gpt(request_text) :
    payload = {
        "messages": [
            {"role": "system", "content": "사용자를 주인님처럼 모셔라"}, 
            {"role": "user", "content": request_text}
        ],
        "max_tokens": 1000,
        "model": "gpt-4o-mini"
    }

    response = requests.post(endpoint, headers=headers, json=payload)

    result_text = ''

    if response.status_code == 200 :
        response_json = response.json()
        result_text = response_json['choices'][0]['message']['content']
    else : 
        print(f'fail request : {response.status_code}')
    
    print('Answer : ', result_text)
```

```python
print('==================GPT랑 채팅하기 ==================')

while True : 
    request_text =  input('질문 입력하세요~!')

    if request_text == 'exit' : 
        break

    request_gpt(request_text)
```

```python
import os

# pip install python-dotenv
from dotenv import load_dotenv

# pip install openai
from openai import AzureOpenAI

def main(): 
        
    try: 
    
        # .env 환경변수 로드
        load_dotenv()
        azure_oai_endpoint = os.getenv("AZURE_OAI_ENDPOINT")
        azure_oai_key = os.getenv("AZURE_OAI_KEY")
        azure_oai_deployment = os.getenv("AZURE_OAI_DEPLOYMENT")
        
        # AzureOpenAI 객체 생성
        client = AzureOpenAI(
            azure_endpoint = azure_oai_endpoint,
            api_key = azure_oai_key,
            api_version="2025-01-01-preview"
        )

        while True:
            # user 입력 메세지
            input_text = input("Enter the prompt (or type 'quit' to exit): ")

            if input_text.lower() == "quit":
                break
            if len(input_text) == 0:
                print("Please enter a prompt.")
                continue # 반복문 최상단으로 이동

            print("Sending request for summary to Azure OpenAI endpoint...")
            
            # Add code to send request...
            messages = [
                {
                    "role": "system",
                    "content": [
                        {
                            "type": "text",
                            "text": "너는 하인이야. 주인님에게 충성하도록!! (극극존칭 사용)"
                        }
                    ]
                },
                {
                    "role": "user",
                    "content": [
                        {
                            "type": "text",
                            "text": input_text
                        }
                    ]
                }
            ]

            print(f'Q : {input_text}')
            
            response = client.chat.completions.create(
                model=azure_oai_deployment, # 사용 모델명(portal에서 확인)
                messages=messages,          # Question  
                max_tokens=6553,            # 모델 응답당 최대 토큰 수(응답 텍스트 최대 길이). 영어는 4글자당 1토큰(모델마다 상이)
                temperature=0.7,            # 0~1 사이 값. 1에 가까울수록 창의성 향상
                top_p=0.95                  # 0~1 사이 값. 모델이 다음 단어 선택 시, 후보 목록의 크기를 조정. 0에 가까울수록 확률이 높은 후보군만 대상이 되므로, 신뢰성 및 정확도 향상
            )
            
            # openai.types.chat.chat_completion.ChatCompletion이라는 별도 객체이므로, .으로 속성 접근 가능
            # 일반 dict는 dict명[속성명] 형식으로 접근 가능
            generated_text = response.choices[0].message.content
            messages.append({"role" : "assistant", "content" : generated_text}) # 이전 응답 정보 저장
            print(f'A : {generated_text}')
            
            

    except Exception as ex:
        print(ex)

# 현재 파일을 직접 실행한경우, __main__이 출력됨.
# 타 모듈에서 호출한경우, 실행되지 않도록 처리.
if __name__ == '__main__': 
    main()
```
